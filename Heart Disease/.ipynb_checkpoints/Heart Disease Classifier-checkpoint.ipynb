Converting categorical variables into dummy variables

```python
dataset = pd.get_dummies(data, columns = ['sex', 'cp', 'fbs','restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

stanScalar = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

#Not Scalaing the Categorical Attributes
dataset[columns_to_scale] = stanScalar.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
X = dataset.drop(['target'], axis=1)
y = dataset['target']
```

## K Neighbors Classifier

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_score = [] # AN array to score the result of each validation
for n in range(1, 21):
    KNN = KNeighborsClassifier(n_neighbors = n)
    score = cross_val_score(KNN, X, y, cv=10)
    knn_score.append(score.mean())
```

```python
knn_score
```

```python
#Finding the value of K corresponding to the maximum accuracy
max=0
for i in range(19):
    if knn_score[max] < knn_score[i]:
        max = i
print(max+1)
```

```python
#Determining best value of K using a  graphical plot
plt.figure(figsize=(15,10))
plt.plot([k for k in range(1,21)], knn_score, color="red")
for i in range(1,21):
    plt.text(i, knn_score[i-1], (i, knn_score[i-1]))
plt.xticks([i for i in range(1,21)])
plt.xlabel('Neighbors(K)')
plt.ylabel('Score')
```

Best Accuracy for K=12

```python
KNN = KNeighborsClassifier(n_neighbors = 12)
score = cross_val_score(KNN, X, y, cv=10)
```

```python
score.mean()
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RN = RandomForestClassifier(n_estimators=12)
score = cross_val_score(RN, X, y, cv=10)
```

```python
score.mean()
```

## Decison Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier(max_depth=6)
score = cross_val_score(DTC, X, y, cv=10)
```
